## Encode and Decode sample 

```python
import os
import numpy as np
from pathlib import Path
from cyber.utils.video import open_video
from notebook_utils import encode_and_save_tokens, decode_tokens_to_video
```

```python
current_directory = Path().resolve()
```

```python
# download ckpt
video_path = os.path.join(current_directory, 'demo_data/test_video.mp4')
ckpt_path = 'path/to/magvit2.ckpt'  # replace with the path to the ckpt file
output_path = f"output/decoded_{os.path.basename(video_path)}"
fps = 30
```

```python
# load video and preprocessing
frames = open_video(video_path)
```

```python
# Encode and save tokens
tokens_file = encode_and_save_tokens(frames, ckpt_path, video_path)
```

```python
# load tokenized data
tokens = np.memmap(os.path.join(current_directory, 'output/tokens_test_video.bin'), dtype=np.uint32, mode='r')
print(tokens.shape)
```

```python
# decode tokens and reconstruct video
decode_frame = decode_tokens_to_video(tokens, ckpt_path, output_path, fps)
```
